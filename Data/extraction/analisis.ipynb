```python
import pandas as pd
import kagglehub
from kagglehub import KaggleDatasetAdapter
import plotly.graph_objects as go
from datetime import datetime
from bs4 import BeautifulSoup
import requests
```

```python
stocks = "Netflix Inc. (NFLX) Stock Price 2002-2025.csv"
IMDB = "Netflix TV Shows and Movies.csv"

netflixStock = kagglehub.load_dataset(
  KaggleDatasetAdapter.PANDAS,
  "samithsachidanandan/netflix-stock-price-2002-2025",
  stocks,
)

netflixStock["Datetimes"] = [datetime.strptime(date_str, '%b %d, %Y') for date_str in netflixStock["Date"]]
netflixStock = netflixStock[netflixStock["Datetimes"] >  datetime(2019, 1, 1)]
netflixStock = netflixStock.sort_values(by="Datetimes")
netflixStock.to_csv("Netflix_Stock.csv", index=False)
fig = go.Figure(data=go.Ohlc(x=netflixStock['Date'],
                    open=netflixStock['Open '],
                    high=netflixStock['High'],
                    low=netflixStock['Low'],
                    close=netflixStock['Close']))
fig.show()
netflixStock.to_csv("Netflix_Stock.csv", index=False)
```

```python
try:
    opinionsDic = {}
    for i in range(1,50):
        url = f"https://www.trustpilot.com/review/www.netflix.com?page={i}"
        html = requests.get(url)
        soup = BeautifulSoup(html.text)
        opinions = soup.find_all("section", class_="styles_reviewContentwrapper__Tzamw")
        for opinion in opinions:
            opinionsContent = opinion.find("p", class_="typography_body-l__v5JLj typography_appearance-default__t8iAq")
            opinionsDate = opinion.find("span", class_="typography_body-m__k2UI7 typography_appearance-subtle__PYOVM")
            opinionsRate = opinion.find("img")
            if opinionsContent is not None:
                type(opinionsContent)
                finalOpinion = {"Content": opinionsContent.text, "Rate": opinionsRate["alt"]}
            else:
                finalOpinion = {"Content": "", "Rate": opinionsRate["alt"]}
            if opinionsDic.keys().__contains__(opinionsDate.text):
                opinionsDic[opinionsDate.text].append(finalOpinion)
            else:
                opinionsDic[opinionsDate.text] = [finalOpinion]
    print("Data retrieved successfully")
    with open("opinionsDic.json", "w") as file:
        file.write(str(opinionsDic))
except Exception as e:
    print("Error at retrieving data")
    print("Error: ",e)
```
